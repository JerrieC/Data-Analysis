```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

#### 1.From the following matrices, find the λ eigenvalues, and the ν eigenvectors as many as possible.

```python
# a.
x=np.array([[0,1,0],[0,0,1],[2,1,-2]])
a,b=np.linalg.eig(x) 
print('λ eigenvalues of this matrix are', a)
print('ν eigenvectors of this matrix are\n', b)
```

```python
# b.
x=np.array([[2,0,0,0],[1,2,0,0],[0,0,0,1],[0,0,-6,5]])
a,b=np.linalg.eig(x) 
print('λ eigenvalues of this matrix are', a)
print('ν eigenvectors of this matrix are\n', b)
```

#### 2.Find the Hubs and Authorities of the graphs below given by HITS. Are the results consistent with the notions of Hubs and Authorities?

```python
# a
# Graph 1
G1 = nx.DiGraph()
G1.add_edges_from([(1,2),(2,4),(4,5),(5,1),(3,1),(3,2),(4,1)])
plt.figure() 
nx.draw(G1 , node_color='c' , with_labels=True, node_size=800)
plt.show()
h1,a1=nx.hits(G1)
print("The Hubs of the graph 1 is ", h1)
print("The Authorities of the graph 1 is ", a1)
```

```python
# b
DG = nx.DiGraph()
DG.add_edges_from([(1,2), (1,3),(2,4),(2,3),(3,1),(3,4),(4,3)])
DG.out_degree(1, weight='weight')

plt.figure()
nx.draw(DG , node_color='c' , with_labels=True, node_size=2000)
plt.show()

h,a=nx.hits(DG)
print('The Hubs of the graph are', h)
print('The Authorities of the graph are', a)
```

#### Are the results consistent with the notions of Hubs and Authorities?
The descreasing order of Hubs is 2->1->4->3, while the order of authorities is 3->4->2->1. Therefore, the results are not consistent with the notions of Hubs and Authorities.
