### c. Bootstrapping

```python
def bootstrap_resample(X, n=None):
    """ Bootstrap resample an array_like
    Parameters
    ----------
    X : array_like
      data to resample
    n : int, optional
      length of resampled array, equal to len(X) if n==None
    Results
    -------
    returns X_resamples
    """
    if isinstance(X, pd.Series):
        X = X.copy()
        X.index = range(len(X.index))
    if n == None:
        n = len(X)
        
    resample_i = np.floor(np.random.rand(n)*len(X)).astype(int)
    X_resample = np.array(X[resample_i])  # TODO: write a test demonstrating why array() is important
    return X_resample
```

### c. Ensemble Models

```python
from sklearn.base import BaseEstimator, RegressorMixin, clone
class AverageWeight(BaseEstimator, RegressorMixin):
    def __init__(self,mod,weight):
        self.mod = mod
        self.weight = weight
        
    def fit(self,X,y):
        self.models_ = [clone(x) for x in self.mod]
        for model in self.models_:
            model.fit(X,y)
        return self
    
    def predict(self,X):
        w = list()
        pred = np.array([model.predict(X) for model in self.models_])
        # for every data point, single model prediction times weight, then add them together
        for data in range(pred.shape[1]):
            single = [pred[model,data]*weight for model,weight in zip(range(pred.shape[0]),self.weight)]
            w.append(np.sum(single))
        return w
```

### a. Idea and methodology

We need to understand our data set in the first place. Try to figure out what each feature means and to what extent these features will affect the sale price. We need to take several steps to explore the data.


#### 1. Single variable analysis

 We only focus on the dependent variable (SalePrice). Try to get the description of the whole sale price like the mean, average, standard deviation to get an overall view of sale price. And try to figure out if we need to transform the original data for better fitting. 

#### 2. Multi-variables analysis

We need to consider each of the independent variables and think of several questions:
If we were the house buyers, would we take this feature into consideration? If we take this feature into consideration, then how much is the importance of this feature?(Basement area and living area are both features affecting our decision. But in most situations, people will put the living area in the first place) ? Dose the information of this feature appears in other features (correlation)? Then we go to next step: try to visualize these features with sale price and find the relationship between the dependent variable. We may use several kinds of charts to help us better understand the data. Sometimes, we cannot get exactly what we want only from one single chart. Then we can build the matrix of correlation to figure reduce features that will be involved into the fitting model. Too many features will result in more time in training and sometimes will cause overfitting.

#### 3. Missing data

 In this part, we also have several important questions: What is the proportion of the missing data? Are the missing values random or in some kind of law? Missing data means the reduction of our sample which will affect our analyzing procedure. Besides, we need to make sure we will not hide truth or make some deviation of the original data. When certain feature has over 15% of missing data, we should delete this feature. Meanwhile, some features with missing value can be replaced with some other correlative features. So we do not need to consider these features, either. And we will combine median value, mode value, average value with stand deviation to fill those missing data.

#### 4. Anomalies

We should find those anomalies and outliers and remove these values from data set. With single variable analysis and multi-variables analysis, we can find those anomalies which will affect our final result.

### Feature engineering

Feature engineering is the most important part. The data and features decide the upper limit of the result while models and algorithms are only methods used to approach this limit.We may use many methods of feature engineering to help us improve the result like transformation(categorical-->numerical), nondimensionalization(standardscaler, minmaxscaler), discretization.

 We can also conduct feature selection and feature expansion in this part. Combine some features or raise the dimension may help us train the data in a more accurate way. Also, we can transform some categorical features into numerical ones for better fitting.

After that, We may choose several models to fit and train the data to get the best result. The result may be decided by the dimension, quality, features of the training data. We should also consider the time of the training process because some models really take long time. We will test from the very beginning like linear regression, logistic regression to some more complicated ones like random forest, decision tree even neural networks till we find a model with a satisfying result.
